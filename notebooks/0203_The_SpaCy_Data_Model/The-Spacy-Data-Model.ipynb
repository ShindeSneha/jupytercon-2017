```python
import os, glob
from IPython.display import display, SVG
import pandas as pd
import spacy
import warnings
import nltk
nltk.download('punkt')

nlp = spacy.load('en')
warnings.filterwarnings('ignore')
```

```python
text = u'that movie was excellent'
nltk.tokenize.word_tokenize(text)
```

```python
doc = nlp(text)
for token in doc:
    print "token: {}".format(token)

print
print "original text: {}".format(doc.text)
```
